```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies.shape  ,credits.shape
```

```python
#As we can see that the ID column in both the dataset is same so we will rename the columnin order to concat the dataset.

credits_column_rename=credits.rename(index=str,columns={'movie_id':'id'})
```

```python
credits_column_rename.head()
```

```python
#Merge the 2 columns

movies_merge_df=movies.merge(credits_column_rename,on='id')

movies_merge_df.head()
```

```python
movies_merge_df.columns
```

```python
movies_clean_df=movies_merge_df.drop(['title_x','title_y','homepage','production_companies','cast'],axis=1)
```

```python
movies_clean_df.head()
```

```python
movies_clean_df.info()
```

```python
#We will use the Weighted Average formula .
v=movies_clean_df['vote_count']
R=movies_clean_df['vote_average']
C=movies_clean_df['vote_average'].mean()
m=movies_clean_df['vote_count'].quantile(0.70)
```

```python
movies_clean_df['Weighted_Average']=((R*v)+ (C*m))/(v+m)
```

```python
movies_clean_df.head()
```

```python
movies_sorted_ranking=movies_clean_df.sort_values(by='Weighted_Average',ascending=False)
```

```python
movies_sorted_ranking[['original_title','Weighted_Average','vote_count','vote_average','popularity']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#The plot shows us the top 10 movies having weighted average nearly equal
sns.barplot(x=movies_sorted_ranking['Weighted_Average'].head(10),y=movies_sorted_ranking['original_title'].head(10),data=movies_sorted_ranking)
plt.show()
```

```python
#Plot graph based on popularity
popularity=movies_clean_df.sort_values(by='popularity',ascending=False)
sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10),data=popularity)
plt.show()
```

Recommendation System Based on Weighted_Average and Popularity

We will give both Weighted_Average and Popularity 50% each importance

```python
movies_clean_df[['original_title','Weighted_Average','popularity']].head()
```

We can see that both Weighted_Average and Popularity have diffrent scale.

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_clean_df[['Weighted_Average','popularity']])
movie_normalize_df=pd.DataFrame(movie_scaled_df,columns=[['Weighted_Average','popularity']])
```

```python
movie_normalize_df.head()
```

```python
movies_clean_df[['Normalized_Weighted_Average','Normalized_Popularity']]=movie_normalize_df
movies_clean_df.head()
```

```python
movies_clean_df['Score']=movies_clean_df['Normalized_Weighted_Average']*0.5+movies_clean_df['Normalized_Popularity']*0.5

movies_clean_df.head()
```

```python
movies_score_df=movies_clean_df.sort_values('Score',ascending=False)
movies_score_df[['original_title','Weighted_Average','popularity','Score']].head()
```

```python
#Plot graph based on Score
#popularity=movies_clean_df.sort_values(by='popularity',ascending=False)
sns.barplot(x=movies_score_df['Score'].head(10),y=movies_score_df['original_title'].head(10),data=movies_score_df)
plt.show()
```
